```python
#unique cities where matches have been played 
data.city.unique()
```

```python
#matches played in each city:
cities = data.groupby('city')[['id']].count()
```

```python
cities
```

```python
cities.rename(columns={'id':'matches'},inplace=True)
cities = cities.sort_values('matches',ascending=True).reset_index()
cities
```

```python
plt.figure(figsize=(20,10))
plt.grid()
plt.title('Number Of Matches Played In Each City')
sns.barplot(x='matches',y='city',data=cities);
plt.savefig('cities.png', dpi=300)
```

```python
#No of matches won by each team:
winner_df = data.groupby('winner')[['id']].count()
winner_df = winner_df.sort_values('id', ascending=False).reset_index()

winner_df.rename(columns = {'id':'wins','winner':'Teams'},inplace=True)
winner_df
```

```python
plt.figure(figsize=(30,20))
plt.xlabel('Teams')
plt.ylabel('Wins')
plt.title('Matches Won By Each Team');
plt.bar(winner_df.Teams,winner_df.wins);
plt.savefig('wins.png', dpi=300)
```

```python
#Matches Played in Each Season:
yearly = data.groupby('Season')[['id']].count()
yearly = yearly.sort_values('id', ascending=False).reset_index()
yearly.rename(columns = {'id':'Matches'},inplace=True)
yearly
```

```python
plt.figure(figsize=(30,20))
plt.xlabel('Year')
plt.ylabel('Matches')
plt.title('Matches Played In Each Season');
plt.bar(yearly.Season,yearly.Matches, color='#bf88be');
plt.savefig('matches.png', dpi=300)
```

```python
data.head()
```

```python
#What was the most preferred Decision On winning Toss i.e. Choose To Bat / Choose To Field

toss_dec = data.groupby('toss_decision')[['id']].count()
toss_dec =toss_dec.sort_values('id', ascending=False).reset_index()
toss_dec.rename(columns = {'id':'Matches','toss_decision':'TossWinnerDecsion'},inplace=True)
toss_dec
```

```python
plt.figure(figsize=(5,5))
plt.xlabel('Toss Winner Decision')
plt.ylabel('Matches')
plt.title('Preferred Toss Decision In Matches');
plt.bar(toss_dec.TossWinnerDecsion,toss_dec.Matches,color=['#4863A0','#566D7E']);
plt.savefig('toss.png', dpi=300)
```

```python
#Which Venue has hosted the Most Number Of Ipl Matches

venues = data.groupby('venue')[['id']].count()
venues =venues.sort_values('id', ascending=False).reset_index()
venues.rename(columns = {'id':'Matches','venue':'Venue'},inplace=True)
venues
```

```python
plt.figure(figsize=(100,70))
plt.xlabel('Venue')
plt.ylabel('Matches')
plt.title('Matches Played In Each Venue');
plt.bar(venues.Venue,venues.Matches, color='#bf88be');
plt.savefig('venues1.png', dpi=300)
```

```python
labels = list(venues.Venue)
plt.figure(figsize=(20,20))
plt.title("Venues",fontweight='bold',fontsize=30)
plt.tick_params(labelsize=40)
plt.pie(venues.Matches,labels=labels,textprops={'fontsize': 13});
plt.savefig('venues2.png', dpi=300)
```

Clearly,Eden Gardens Has Hosted Maximum number of IPL Matches

```python
#Who has been awarded with Player Of the Max maximum Number Of Times

players = data.groupby('player_of_match')[['id']].count()
players =players.sort_values('id', ascending=False).reset_index()
players.rename(columns = {'id':'Awards'},inplace=True)
players

labels = list(players.player_of_match)
plt.figure(figsize=(20,20))
plt.title("Awards",fontweight='bold',fontsize=30)
plt.tick_params(labelsize=40)
plt.pie(players.Awards,labels=labels,textprops={'fontsize': 13});
plt.savefig('awards.png', dpi=300)
```

```python
plt.figure(figsize=(100,70))
plt.xlabel('Player Of The Match')
plt.ylabel('Awards')
plt.title('Awards');
plt.xticks(rotation=90)
plt.yticks(ticks=np.arange(0,25,5))
plt.bar(players.player_of_match,players.Awards, alpha=0.6);
plt.savefig('player_of_match.png', dpi=300)
```

From the above result it is clear that Chris Gayle has received  "21 Man of The Match Titles" and is followed by AB de Villiers having "20"

```python
# Who Has Won the Ipl Trophy Most Number of Times
teams=data.groupby('Season').tail(1).copy()
teams =teams.sort_values('id', ascending=False).reset_index()
teams
```

```python
teams.winner.unique()
```

```python
plt.figure(figsize=(100,70))
plt.xlabel('Team')
plt.ylabel('IPL Titles')
plt.title('Teamwise IPL Titles');
plt.bar(teams.winner,teams.id, color='#bf88be');
plt.savefig('IPL_titles.png', dpi=300)
```

Clearly,Mumbai Indians Have Won Most Number Of IPL Titles
